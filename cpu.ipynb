```python
import pandas as pd
import numpy as np
data = pd.read_html("https://www.techpowerup.com/cpu-specs/?mobile=No&server=No&igp=Yes&sort=name")[0]
```

```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
filt = data.iloc[:,[0,2,3,4,5,6,7,8]]
filt.columns = ['Name','cores','clock','socket','nm','L3','TDP',"data"]
filt.loc[:,'TDP'] = (filt.loc[:,'TDP'].map(lambda x : int(x.split(" ")[0]))).copy()
filt.loc[:,'nm'] = (filt.loc[:,'nm'].map(lambda x : int(x.split(" ")[0]))).copy()
filt.loc[:,'cores'] = (filt.loc[:,'cores'].map(lambda x : int(x.split("/")[1] if x.find("/") != -1 else x))).copy()
filt.sort_values(by = ["TDP",'cores','nm'],ignore_index = True)
```
