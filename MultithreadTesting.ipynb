```python
import time
import threading
import queue

# Worker 類別，負責處理資料
class Worker(threading.Thread):
    def __init__(self, queue, num):
        threading.Thread.__init__(self)
        self.queue = queue
        self.num = num

    def run(self):
        while self.queue.qsize() > 0:
            # 取得新的資料
            msg = self.queue.get()
            test(msg)
            # 處理資料
            print(f"\nWorker {self.num}:{type(msg)}:{msg} ")
            time.sleep(3)

# 建立佇列
my_queue = queue.Queue()
array = list(range(1,100))
def test(i):
    array[i] = i*i
# 將資料放入佇列
for i in range(100):
    my_queue.put(i)

# 建立兩個 Worker
Workers = []
for x in range(100):
    Workers.append(Worker(my_queue, x + 1))
    # 讓 Worker 開始處理資料
    Workers[x].start()
    
# 等待所有 Worker 結束
for x in range(100):
    Workers[x].join()


print("Done.")
array
```
